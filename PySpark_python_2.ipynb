```python
from pyspark.sql import SparkSession
spark=SparkSession.builder.appName('Practise').getOrCreate()
```

```python
df = spark.read.csv('/content/sample_data/california_housing_test.csv',header=True,inferSchema=True)
```

```python
df.show()
```

```python
df.printSchema()
```

```python
df.describe().show()
```

```python
df.na.drop().show()
```

```python
df.na.drop(how='all').show()
```

```python
df.na.drop(how='any').show()
```

```python
df.na.drop(thresh=2).show()
```

```python
df.na.drop(subset=['longitude']).show()
```

```python
df.na.fill('Missing Values',['longitude']).show()
```

```python
df.na.fill(0,['longitude']).show()
```

```python
df.show()
```

```python
from pyspark.ml.feature import Imputer
imputer = Imputer(
    inputCols=['longitude','total_rooms'],
    outputCols=["{}_imputed".format(c) for c in ['longitude','total_rooms']]
).setStrategy("mean")
```

```python
imputer.fit(df).transform(df).show()
```
