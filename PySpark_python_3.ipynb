```python
from pyspark.sql import SparkSession
saprk = SparkSession.builder.appName('practice').getOrCreate()
```

```python
df=saprk.read.csv('/content/sample_data/california_housing_train.csv',header=True,inferSchema=True)
```

```python
df.show()
```


Pyspark Dataframes

*   Filter Operation
*   &,|,==
*   ~


```python
df.filter("housing_median_age<10").show()
```

```python
df.filter("total_rooms == 2000").select(['median_income']).show()
```

```python
df.filter(df['housing_median_age']<10).show()
```

```python
df.filter((df['housing_median_age']<10) & (df['total_rooms']==2000)).show()
```

```python
df.filter(~(df['housing_median_age']<10)).show()
```

```python
df.groupBy('housing_median_age').count().show()
```

```python
df.groupBy('housing_median_age').mean().show()
```

```python
df.groupBy('housing_median_age').max().show()
```

```python
df.filter((df['housing_median_age']<10 )| (df['total_rooms']==2000)).show()
```

```python
df.filter((df['housing_median_age']<10 )& ~(df['total_rooms']==2000)).show()
```
